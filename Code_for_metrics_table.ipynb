```python
from google.colab import files
import io
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
uploaded = files.upload()
```

```python
dataset_df = pd.read_csv(io.BytesIO(uploaded['datatfg.csv']))
```

```python
import pandas as pd
from IPython.display import display, HTML


pd.set_option('display.max_columns', None)

description = dataset_df.describe()

styled_table = description.style.set_table_attributes('class="dataframe"') \
    .set_properties(**{'border-collapse': 'collapse', 'border': '1.5px solid black'}) \
    .set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#b04c4c'),
                                     ('color', 'white'),
                                      ('border', '1.5px solid black'),
                                      ('text-align', 'center')]},
        {'selector': 'td', 'props': [('border', '1px solid black')]},
        {'selector': 'caption', 'props': [('caption-side', 'top')]},
    ])

display(HTML(styled_table.render()))

styled_table_html = styled_table.render()
with open('styled_table.html', 'w') as f:
    f.write(styled_table_html)
```
